```python
!pip install Pandas
!pip install Seaborn

import  pandas as pd
import seaborn as sb

df = pd.read_csv("iris.csv")
```

```python
df.head()
df.describe()
df.replace(
 {"target":
  {0: "Iris-setosa",
   1: "Iris-versicolor",
   2: "Iris-virginica"}},
 inplace=True)
df.to_csv("changed_iris.csv", index=False)
df.head()
```

```python
sb.set_theme(style="white")
sb.color_palette("tab10")
g = sb.PairGrid(df, diag_sharey=False, hue="target")
g.map_upper(sb.scatterplot, s=15)
g.map_lower(sb.kdeplot)
g.map_diag(sb.kdeplot, lw=2)
g.add_legend()
```
